```python
#file is utf8 as utf16 so null characters everywhere
# remove null characters w/ following:
#sed -i 's/\x0//g' diagnostics_map1_evts.log

import csv
import re
import numpy as np

filepath="/home/lachlan/CODEBASE/ReadoutXFM/data/diagnostics_map1_evts.log"

with open(filepath) as f:
    nlines_init=sum(1 for line in f)
    print(nlines_init)
    f.seek(0)
    rt=np.zeros(nlines_init, dtype=float)
    lt=np.zeros(nlines_init, dtype=float)
    tr=np.zeros(nlines_init, dtype=int)
    ev=np.zeros(nlines_init, dtype=int)
    icr=np.zeros(nlines_init, dtype=float)
    ocr=np.zeros(nlines_init, dtype=float)

    print(rt.shape, rt[1], nlines_init)

    nlines=0
    npx=0
    nmaps=0

    for line in csv.reader(f):
        if "Map Acquire start" in line[0]:
            print(f"Map started at line: {nlines}")
            nmaps += 1
            if nmaps > 1:
                raise ValueError(f"Multiple maps in file, additional map starting at line: {nlines}")
        if "Deadtime realtime" in line[0]:
            print(line)
            rt[npx] = float(re.findall("[-+]?[.]?[\d]+(?:,\d\d\d)*[\.]?\d*(?:[eE][-+]?\d+)?", line[0][8:])[0])
            lt[npx] = float(re.findall("[-+]?[.]?[\d]+(?:,\d\d\d)*[\.]?\d*(?:[eE][-+]?\d+)?", line[1])[0])
            tr[npx ]= int(re.findall("[-+]?[.]?[\d]+(?:,\d\d\d)*[\.]?\d*(?:[eE][-+]?\d+)?", line[2])[0])
            ev[npx] = int(re.findall("[-+]?[.]?[\d]+(?:,\d\d\d)*[\.]?\d*(?:[eE][-+]?\d+)?", line[3])[0])
            icr[npx ]= float(re.findall("[-+]?[.]?[\d]+(?:,\d\d\d)*[\.]?\d*(?:[eE][-+]?\d+)?", line[4])[0])
            ocr[npx] = float(re.findall("[-+]?[.]?[\d]+(?:,\d\d\d)*[\.]?\d*(?:[eE][-+]?\d+)?", line[5])[0])
            print(rt[npx], lt[npx], tr[npx], ev[npx], icr[npx], ocr[npx])
            npx+=1
        nlines+=1    

    rt=rt[:npx]
    lt=lt[:npx]
    tr=tr[:npx]
    ev=ev[:npx]
    icr=icr[:npx]
    ocr=ocr[:npx]

print(f"last values: {rt[-1]} {lt[-1]} {tr[-1]} {ev[-1]} {icr[-1]} {ocr[-1]}")

print(f"lines found: {nlines_init}, lines read: {nlines}, pixels: {npx}, stored: {len(rt)}")
```
