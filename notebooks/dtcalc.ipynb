```python
import numpy as np
import codecs
import matplotlib.pyplot as plt

XDIM=256
YDIM=107
ndet = 2

with codecs.open('../out/ts2_0_export/pixeldata/pxstats_dt.txt') as f1:
    ypred = np.genfromtxt(f1, delimiter=',')

with codecs.open('../out/ts2_original/pixeldata/pxstats_dt.txt') as f2:
    y = np.genfromtxt(f2, delimiter=',')

```

```python
yavg=(y[0,:]+y[1,:])/2

yavg = np.delete(yavg, 0)
yavg = np.append(yavg, 0)
```

```python
print(f"{ypred.shape}, {yavg.shape}")
```

```python
diffmap = ypred-yavg

diffimage = diffmap.reshape(YDIM,XDIM)

fig = plt.figure(figsize=(12,6))

ax = fig.add_subplot(111)

img = ax.imshow(diffimage, cmap='bwr')

plt.colorbar(img, fraction=0.04346, pad=0.04)
```
